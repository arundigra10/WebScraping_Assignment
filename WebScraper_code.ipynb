```python
import re
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.marketsandmarkets.com/telecom-and-IT-market-research-113.html"
response = requests.request("GET", url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
div = soup.find("div", {"id": "publishedreports"})
type(div)
```

### Report Titles Scraping

```python
report_titles=[]
for link in div.find_all('a'):
    report_titles.append(link.text)
    print(link.text)
```

### Date Published

```python
date_pub=[]
i=0
for date in div.find_all('td',"displaynone","center"):
   if i%3==0: 
     date_pub.append(date.text.strip().replace("\xa0", " "))
     print(date.text.strip().replace("\xa0", " "))
   i+=1
```

```python
print(date_pub)
```

### Link of each report

```python
report_links=[]
for link in div.find_all('a'):
    report_links.append("https://www.marketsandmarkets.com/"+link.get('href'))
    print("https://www.marketsandmarkets.com/"+link.get('href'))
```

### Report descriptions

```python
reprot_description=[]
for messi in div.find_all('p',{"align": "justify"}):
    barca=re.search(r".+?[^0-9]\.",messi.text)
    reprot_description.append(barca.group())
    print(barca.group()) 
```

### USD values

```python
usd_values=[]
for messi in div.find_all('p',{"align": "justify"}):
    barca=re.search(r"USD [0-9.]+ [BbMm]illion",messi.text)
    usd_values.append(barca.group())
    print(barca.group()) 
```

### % Val

```python
percent_val=[]
for messi in div.find_all('p',{"align": "justify"}):
    fifa=re.search(r"[0-9.\s]+%",messi.text)
    percent_val.append(fifa.group())
    print(fifa.group()) 
```

```python
print(f"{len(report_titles)} {len(date_pub)} {len(report_links)} {len(reprot_description)} {len(usd_values)} {len(percent_val)}  ")
```

### Scraped Data to CSV File

```python
df=pd.DataFrame()
df["report_title"]=report_titles
df["date_published"]=date_pub
df["link"]=report_links
df["report_description"]=reprot_description
df["USD _value"]=usd_values
df["%_val"]=percent_val
df.to_csv("skyquest_assignment.csv", index=False)
```
